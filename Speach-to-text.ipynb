```python
from googletrans import Translator
import gradio as gr
```

```python


# Load the Whisper ASR pipeline
whisper = pipeline('automatic-speech-recognition', model='openai/whisper-medium', device=0)

def transcribe_and_translate(audio_file, target_language):
    try:
        transcription = None

        audio, rate = torchaudio.load(audio_file.name)

        if audio.shape[0] == 2:
            audio = audio.mean(dim=0, keepdim=True)

        target_sample_rate = 16000
        if rate != target_sample_rate:
            resampler = torchaudio.transforms.Resample(orig_freq=rate, new_freq=target_sample_rate)
            audio = resampler(audio)

        transcription = whisper(audio[0].numpy())

        transcribed_text = transcription['text']

        translator = Translator()

        # Define supported languages and their names for the dropdown menu
        supported_languages = {
            'en': 'English',
            'ur': 'Urdu',
            'fr': 'French',
            'es': 'Spanish',
            'de': 'German',
            'it': 'Italian',

        }

        # Check if the target_language is a valid language code
        if target_language in supported_languages:
            translation = translator.translate(transcribed_text, src='en', dest=target_language)
            translated_language = supported_languages[target_language]
            return transcribed_text, f"Translated to {translated_language}: {translation.text}"
        else:
            return "Invalid target language selection. Please choose a language from the dropdown menu.", ''
    except Exception as e:
        return str(e), ''


```

```python
# Define the dropdown menu for selecting the target language
language_choices = [
    ('English', 'en'),
    ('Urdu', 'ur'),
    ('French', 'fr'),
    ('Spanish', 'es'),
    ('German', 'de'),
    ('Italian', 'it'),

]

iface = gr.Interface(
    fn=transcribe_and_translate,
    inputs=[
        gr.inputs.File(label="Upload an audio file"),
        gr.inputs.Radio(label="Select Target Language", choices=language_choices)
    ],
    outputs=[gr.outputs.Textbox(label="Transcribed Text"), gr.outputs.Textbox(label="Translated Text")],
)

iface.launch(share=True)
```
